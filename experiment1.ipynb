```python
#Task 2: Perform data acquisition and exploratory analysis using descriptive statistical measures to understand dataset characteristics
#(Suggested Dataset: Pima Indians Diabetes Dataset / Iris Dataset).
```

```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
iris_raw = load_iris()
```

```python
iris_df = pd.DataFrame(data=iris_raw.data, columns=iris_raw.feature_names)
```

```python
iris_df
```

```python
iris_df['species_id'] = iris_raw.target
```

```python
species_mapping = {i: name for i, name in enumerate(iris_raw.target_names)}
```

```python
iris_df['species_name'] = iris_df['species_id'].map(species_mapping)
```

```python
print("Iris dimensions:", iris_df.shape)
```

```python
print("Missing entries checking:\n", iris_df.isnull().sum())
```

```python
try:
    unique_species = iris_df['species_id'].unique()
    print(unique_species)
except KeyError:
    print("Error: Column 'species_id' not found in DataFrame.")
```

```python
iris_df.head()
```

```python
for col in iris_raw.feature_names:
    mean_val = iris_df[col].mean()
    median_val = iris_df[col].median()
    mode_val = iris_df[col].mode()[0]
    print(f"Feature: {col}")
    print(f"  Mean:   {mean_val:.4f} cm")
    print(f"  Median: {median_val:.4f} cm")
    print(f"  Mode:   {mode_val:.4f} cm")
    print("-" * 40)
```

```python
for col in iris_raw.feature_names:
    variance = iris_df[col].var()
    std_dev = iris_df[col].std()
    data_range = iris_df[col].max() - iris_df[col].min()
    q1 = iris_df[col].quantile(0.25)
    q3 = iris_df[col].quantile(0.75)
    print(f"Feature: {col}")
    print(f"  Variance:           {variance:.4f}")
    print(f"  Std Deviation:      {std_dev:.4f}")
    print(f"  Range (Max - Min):  {data_range:.4f}")
    print(f"  IQR (Q3 - Q1):      {q3 - q1:.4f}")
    print("-" * 40)
```

```python
for col in iris_raw.feature_names:
    skewness = iris_df[col].skew()
    kurtosis = iris_df[col].kurt()
    print(f"Feature: {col}")
    print(f"  Skewness (Asymmetry): {skewness:.4f}")
    print(f"  Kurtosis (Tails):     {kurtosis:.4f}")
    print("-" * 40)
```

```python
#Task 3: Analyze data distributions and feature relationships using univariate and multivariate visualization techniques (Suggested Dataset: Pima Indians Diabetes Dataset / Iris Dataset).
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris Dataset
df = sns.load_dataset("iris")

# Display first five rows
print(df.head())
```

```python
# Histogram
plt.figure(figsize=(6,4))
sns.histplot(df["sepal_length"], bins=15, kde=True)
plt.title("Histogram of Sepal Length(24eu01076)")
plt.savefig('histogram')
plt.show()
```

```python
# Count Plot
plt.figure(figsize=(6,4))
sns.countplot(x="species", data=df)
plt.title("Species Count(24eu01076)")
plt.savefig('count')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(6,4))
sns.boxplot(y="petal_length", data=df)
plt.title("Box Plot of Petal Length(24eu01076)")
plt.savefig('box')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(6,4))
sns.violinplot(y="petal_width", data=df)
plt.title("Violin Plot of Petal Width(24eu01076)")
plt.savefig('violin')
plt.show()
```

```python
# Density Plot
plt.figure(figsize=(6,4))
sns.kdeplot(df["sepal_width"], fill=True)
plt.title("Density Plot of Sepal Width(24eu01076)")
plt.savefig('density')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(6,4))
sns.scatterplot(x="sepal_length", y="petal_length",
                hue="species", data=df)
plt.title("Sepal Length vs Petal Length(24eu01076)")
plt.savefig('scatter')
plt.show()
```

```python
# Pair Plot
sns.pairplot(df, hue="species")
plt.show()
plt.savefig('pairplot')
```

```python
# Correlation Heatmap
plt.figure(figsize=(7,5))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")
plt.title("Correlation Heatmap(24eu01076)")
plt.savefig('correlation')
plt.show()
```

```python
# Box Plot by Species
plt.figure(figsize=(6,4))
sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Petal Length by Species(24eu01076)")
plt.savefig('petal length')
plt.show()
```

```python
# Violin Plot by Species
plt.figure(figsize=(6,4))
sns.violinplot(x="species", y="sepal_width", data=df)
plt.title("Sepal Width by Species(24eu01076)")
plt.savefig('sepal width')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(6,4))
sns.barplot(x="species", y="petal_width", data=df)
plt.title("Average Petal Width by Species(24eu01076)")
plt.savefig('average')
plt.show()
```

```python
import pandas as pd

# Load Sonar Dataset from GitHub
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/sonar.csv"

df = pd.read_csv(url, header=None)

# Rename the last column as Target
df.rename(columns={60: "Target"}, inplace=True)

print(df.head())
```
