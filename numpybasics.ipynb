```python
import numpy as np
```

```python

numbers = np.array([10, 20, 30, 40])
print(numbers)  
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix)
```

```python
prices = np.array([100, 250, 500])
discounted_prices = prices * 0.90
print(discounted_prices)
bonus = np.array([5, 10, 15])
final_prices = discounted_prices + bonus
print(final_prices)
```

```python
scores = np.array([85, 92, 78, 99, 88])
print(np.mean(scores))  
print(np.max(scores))       
print(np.min(scores))       
```

```python
ages = np.array([14, 22, 17, 35, 41, 12, 19])
is_adult = ages >= 18
adult_ages = ages[is_adult]
print(adult_ages)
```

```python
flat_data = np.array([1, 2, 3, 4, 5, 6])
matrix = flat_data.reshape(2, 3)
print(matrix)
```

```python
numbers = np.array([15, 3, 99, 42, 1])
sorted_numbers = np.sort(numbers)
print(sorted_numbers)
```

```python
salaries = np.array([2500, 3100, np.nan, 4000])
clean_average = np.nanmean(salaries)
print(clean_average)
```

```python
data = np.array([10, np.nan, 30, np.nan, 50])
clean_data = data[~np.isnan(data)]
print(clean_data)  
data[np.isnan(data)] = 0
print(data)        
```

```python
box = np.array([[1, 2, 3], [4, 5, 6]])
print(box.size)  
print(box.shape) 
```
